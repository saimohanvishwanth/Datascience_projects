# Histograms

```python
#Histogram:
'''steps to be taken:1.bin the range of the values
    2.divide the entire range of values in to series of intervals
    3.count how many values fall in each interval
    matplotlib.pyplot.hist()'''
import matplotlib.pyplot as plt
import numpy as np
fig, ax=plt.subplots(1,1)
a=np.array([34,56,1,23,42,3,2,34,1,12,12,14,21,2])
ax.hist(a,bins=[0,20,30,40,50,70,80,90,100])
ax.set_title('histogram of result')
ax.set_xticks([0,20,25,50,75,100])
ax.set_xlabel('marks')
ax.set_ylabel('no of students')
plt.show()
```

# pie chart

```python
data=[123445,543326,234254,683754,342178,234216]
l=['domins','mfc','kfc','mcdonalds','burgerking','rollercoster']
e=[0,0.0,0,0.0,0,0.0]
colors=['black','orange','blue','white','green','indigo']
plt.pie(data, labels=l, explode=e,shadow=False,startangle=90,wedgeprops= {'edgecolor':'black', 'linewidth':1},colors=colors)
plt.title('population pie chart')
plt.grid(True)
plt.show()
```

# scatter plot

```python
x=[34,5.6,7,23,89,12,23,86]
y=[98,4.56,12.5,34,85,34,93,67.32]

x1=[64.2,84.6,97.5,86.8,76.3,96.1,74.5]
y1=[23,54,75,86,97,23,52]

plt.scatter(x,y,label='high income low saving',color='r')
plt.scatter(x1,y1,label='low income high saving',color='b')
plt.xlabel('high 100')
plt.ylabel('high saving 1000')
plt.title('expenditure and saving on common people living in INDIA')
plt.legend()
plt.show()
```
